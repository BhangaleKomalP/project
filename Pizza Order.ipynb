# PIZZA ORDER AND THEIR BILL

```python
l=[]
l.append(input('Enter Your Name:- '))
l.append(input('Enter Your 10 digit Mobile No.:- '))
def Order_details():
    main_order={}
    while True:
        print('''choose one option
        1. For veg pizza's
        2. For non-veg 
        3. To view order list
        4. To edit order list
        5. To quit''')
        choice=input('Enter your choice = ')
        if choice=='1':
            while True:
                print('''
        1. Cheese pizza         :- Rs.250/pizza
        2. Paneer pizza         :- Rs.200/pizza
        3. Margherita pizza     :- Rs.150/pizza
        4. Go to main menu''')
                choice1=int(input('Enter your pizza code = '))
                if choice1==1:
                    number=int(input('Enter how many Cheese pizza you want = '))
                    main_order.setdefault('Cheese pizza     ',number)
                    print(f'\t {number} Cheese pizza has been added')
                elif choice1==2:
                    number=int(input('Enter how many Paneer pizza you want = '))
                    main_order.setdefault('Paneer pizza     ',number)
                    print(f'\t {number} Paneer pizza has been added')
                elif choice1==3:
                    number=int(input('Enter how many Margherita pizza you want = '))
                    main_order.setdefault('Margherita pizza',number)
                    print(f'\t {number} Margherita pizza has been added')
                    
                elif choice1==4:
                    break

                else:
                    print('Invalid choice')

        elif choice=='2':
            while True:
                print('''
        1. Chiken Sausage            :- Rs.300/pizza
        2. Pepper Barbecue chiken    :- Rs.350/pizza
        3. Chiken Dominator          :- Rs.350/pizza
        4. go to main menu''')
                choice2=int(input('Enter your code = '))
                if choice2==1:
                    number=int(input('Enter how many Chiken Sausage you want = '))
                    main_order.setdefault('Chiken Sausage     ',number)
                    print(f'\t {number} Chiken Sausage has been added')

                elif choice2==2:
                    number=int(input("Enter how many Pepper Barbecue chiken you want = "))
                    main_order.setdefault('Pepper Barbecue chiken',number)
                    print(f'\t {number} Pepper Barbecue chiken has been added')
                    
                elif choice2==3:
                    number=int(input("Enter how many Chiken Dominator you want = "))
                    main_order.setdefault('Chiken Dominator',number)
                    print(f'\t {number} Chiken Dominator has been added')
                elif choice2==4:
                    break
                else:
                    print('Invalid choice')
        elif choice=='3':
            print(main_order)
        elif choice=='4':
            print('''
        1. To update order
        2. To delete some items from list
        3. To delete all order''')
            choice4=int(input('Enter your choice = '))
            if choice4==1:
                name=input('Enter what you want to update = ')
                if name in main_order.keys():
                    main_order[name]=int(input('Enter the quantity = '))
                else:
                    print(name,'not available in your order')
            elif choice4==2:
                main_order.pop(input('Enter what you want to delete :-'))
            elif choice4==3:
                main_order.clear()
        elif choice=='5':
            break
        else:
            print('Invalid choice')
    return main_order

def Billing():
    print('NAME:- ',l[0])
    print('Mobile No.:- ',l[1],'\n')
    TOTAL=0
    costs={'Chiken Sausage     ':300,'Pepper Barbecue chiken':350,'Chiken Dominator':350,'Cheese pizza     ':250,'Paneer pizza     ':200,'Margherita pizza':150}
    print('NAME\t\t\t\t(QTY)\t\t COST')
    for i in Main_order:
        print(i,"\t\t(Qty",Main_order[i],')',end='')
        print('\t',costs[i]*Main_order[i])
        TOTAL=TOTAL+costs[i]*Main_order[i]
    print('------------------------------------------------------')
    print(
        '\nTotal\t\t\t\t\t\t',float(TOTAL),
        '\nCGST\t\t\t\t\t\t',TOTAL*0.09,
        '\nSGST\t\t\t\t\t\t',TOTAL*0.09,
        '\nTOTAL COST\t\t\t\t\t',TOTAL+2*(TOTAL*0.09)
    )
    print("\nThank You for your Order\nPlease visit again.....")
```
